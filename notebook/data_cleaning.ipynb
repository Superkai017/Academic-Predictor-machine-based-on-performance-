```python
import pandas as pd
import numpy as np 
data = r"C:\Users\MSI\ML-final-project\Academic-Predictor-machine-based-on-performance-\data\raw_data\cambodian_student_Study_habit_and_thier_performance - Academic_Predictor_Extended.csv"
df = pd.read_csv(data)
```

```python

import re

def extract_midpoint(value):
    if pd.isna(value):
        return None
    
    value = str(value).strip().lower()

   
    single = re.match(r'^(\d+(?:\.\d+)?)\s*(hours?|days?|%)?$', value)
    if single:
        return float(single.group(1))

    
    if "less than 1" in value:
        return 0.5

  
    less = re.match(r'less than (\d+)', value)
    if less:
        return float(less.group(1)) - 1

    plus = re.match(r'(\d+)\+', value)
    if plus:
        return float(plus.group(1))

    range_match = re.search(r'(\d+(?:\.\d+)?)\s*[–\-]\s*(\d+(?:\.\d+)?)', value)
    if range_match:
        lo = float(range_match.group(1))
        hi = float(range_match.group(2))
        return (lo + hi) / 2

    return None  
range_cols = [
    "Self_Study_Hrs",
    "Tutoring_Hrs_Wk",
    "Social_Media_Hrs",
    "Gaming_Hrs",
    "Exercise_Days_Wk",
    "Sleep_Hrs",
]

for col in range_cols:
    df[col] = df[col].apply(extract_midpoint)

def parse_attendance(value):
    if pd.isna(value):
        return None
    value = str(value).strip().lower().replace('%', '')
    if "below 60" in value:
        return 55.0
    range_match = re.search(r'(\d+)\s*[–\-]\s*(\d+)', value)
    if range_match:
        return (float(range_match.group(1)) + float(range_match.group(2))) / 2
    return None

df["Attendance_Rate"] = df["Attendance_Rate"].apply(parse_attendance)
def parse_stress(value):
    if pd.isna(value):
        return None
    range_match = re.search(r'(\d+)\s*[–\-]\s*(\d+)', str(value))
    if range_match:
        return (float(range_match.group(1)) + float(range_match.group(2))) / 2
    return None

df["Stress_Level"] = df["Stress_Level"].apply(parse_stress)


numeric_cols = ["Stress_Level", "Self_Study_Hrs", "Tutoring_Hrs_Wk",
                "Social_Media_Hrs", "Gaming_Hrs", "Exercise_Days_Wk",
                "Sleep_Hrs", "Attendance_Rate"]

print(df[numeric_cols].dtypes)
print("\nSample output:")
print(df[numeric_cols].head(10).to_string())
print("\nMissing values:")
print(df[numeric_cols].isnull().sum())
```

```python
def gpa_to_tier(gpa):
    if pd.isna(gpa):
        return None
    elif gpa < 2.5:
        return "At-Risk"
    elif gpa < 3.0:
        return "Fair / Average"
    elif gpa < 3.5:
        return "Very Good / Good"
    else:
        return "Outstanding / Excellent"

df["Performance_Tier"] = df["GPA"].apply(gpa_to_tier)  # ✅ this is all you need

print("Tier distribution:")
print(df["Performance_Tier"].value_counts())

print("\nSample check:")
print(df[["GPA", "Performance_Tier"]].sort_values("GPA").head(100).to_string())
```
