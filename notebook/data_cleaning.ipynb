```python
import pandas as pd
data = 'Predicting Academic Achievement in Cambodian Higher Education.csv'
df = pd.read_csv(data)
```

```python
df["GPA"].fillna(df["GPA"].median(), inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().any()
df.drop_duplicates(keep='first', inplace=True)
df.shape
```

```python
df['GPA']=df['GPA'].fillna(df['GPA'].median(), inplace=True)
df.isnull().sum()
```

```python
def gpa_to_tier(gpa):
    if pd.isna(gpa):
        return None
    elif gpa < 2.5:
        return "At-Risk"
    elif gpa < 3.0:
        return "Fair / Average"
    elif gpa < 3.5:
        return "Very Good / Good"
    else:
        return "Outstanding / Excellent"

df["Performance_Tier"] = df["GPA"].apply(gpa_to_tier)  # ✅ this is all you need

print("Tier distribution:")
print(df["Performance_Tier"].value_counts())

print("\nSample check:")
print(df[["GPA", "Performance_Tier"]].sort_values("GPA").head(100).to_string())
```

```python
df['Study_Habit'].replace('No fixed habit','Irregular study', inplace=True)
```

```python
df['Year'] = df['Year'].replace('postgraduate education', 'Year 4',inplace=True)
df['Year'].unique()
```

```python
# ── 1. Diet_Quality — normalise casing ───────────────────────────────────────
df["Diet_Quality"] = df["Diet_Quality"].str.strip().str.title()
# "good" → "Good", "excellent" → "Excellent", etc.
 
# ── 2. Internet_Quality — normalise casing ───────────────────────────────────
df["Internet_Quality"] = df["Internet_Quality"].str.strip().str.title()
 
# ── 3. University — strip + title to merge the two RUPP spellings ────────────
df["University"] = df["University"].str.strip().str.title()
 
# ── 4. Commute_Time — unify the two ">1 hour" labels ────────────────────────
df["Commute_Time"] = df["Commute_Time"].replace(
    {"More than 1 hour": "Over 1 hour"}
)
```
