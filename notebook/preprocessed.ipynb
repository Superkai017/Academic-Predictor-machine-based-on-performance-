Feature engineer

```python
commute_map = {
    'Under 15 minutes': 0.125,  # midpoint of 0–15 min
    '15–30 minutes':    0.375,  # midpoint of 15–30 min
    '31–60 minutes':    0.75,   # midpoint of 31–60 min
    'Over 1 hour':      1.25    # reasonable estimate for 60–90 min
}
study_group_map = {
    'Never':                              0,
    'Rarely (Before exams only)':         1,
    'Sometimes (Weekly)':                 2,
    'Frequently (Multiple times a week)': 3
}
study_habit_map = {
    'Cramming':        1,
    'Irregular study': 2,
    'Regular study':   3
}

commute_hrs = df['Commute_Time'].map(commute_map)
study_group_num = df['Study_Group_Freq'].map(study_group_map)
study_habit_num = df['Study_Habit'].map(study_habit_map)
extracurricular_num = df['Extracurricular'].map({'Yes': 1, 'No': 0})

# --- Feature 1: Total Productive Hours ---
df['Total_Productive_Hrs'] = df['Self_Study_Hrs'] + df['Tutoring_Hrs_Wk']

# --- Feature 2: Distraction Hours ---
df['Distraction_Hrs'] = df['Social_Media_Hrs'] + df['Gaming_Hrs']

# --- Feature 3: Study-to-Distraction Ratio ---
df['Study_Distraction_Ratio'] = df['Self_Study_Hrs'] / (df['Distraction_Hrs'] + 0.1)  # +0.1 avoid div by zero

# --- Feature 4: Wellbeing Score ---
df['Wellbeing_Score'] = df['Sleep_Hrs'] + df['Exercise_Days_Wk'] - df['Stress_Level']

# --- Feature 5: Available Study Time ---
df['Available_Study_Time'] = 24 - df['Sleep_Hrs'] - df['Social_Media_Hrs'] - df['Gaming_Hrs'] - commute_hrs

# --- Feature 6: Active Engagement Score ---
df['Active_Engagement_Score'] = df['Attendance_Rate'] / 10 + study_group_num + extracurricular_num

# --- Feature 7: Study Quality Flag ---
df['Study_Quality'] = study_habit_num * df['Self_Study_Hrs']

print(df[['Total_Productive_Hrs', 'Distraction_Hrs', 'Study_Distraction_Ratio',
          'Wellbeing_Score', 'Available_Study_Time', 'Active_Engagement_Score',
          'Study_Quality']].head())
```

```python
df 
```

```python
# Drop high cardinality & leakage
df.drop(columns=['Age', 'University', 'Hometown'], inplace=True)

# Binary
df['Gender']     = df['Gender'].map({'Male': 1, 'Female': 0})
df['STEM_Major'] = df['STEM_Major'].map({'Yes': 1, 'No': 0})

# Ordinal
df['Employment'] = df['Employment'].map({
    'Unemployed / Full-time student': 0,
    'Working part-time':              1,
    'Working full-time':              2
})
df['Diet_Quality'] = df['Diet_Quality'].map({
    'Poor': 1, 'Average': 2, 'Good': 3, 'Excellent': 4
})
df['Internet_Quality'] = df['Internet_Quality'].map({
    'Poor': 1, 'Fair': 2, 'Good': 3, 'Excellent': 4
})
df['AI_Tool_Usage'] = df['AI_Tool_Usage'].map({
    'Never': 0, 'Rarely': 1, 'Sometimes': 2, 'Often': 3, 'Always': 4
})
df['Year'] = df['Year'].map({
    'Year 1': 1, 'Year 2': 2, 'Year 3': 3, 'Year 4': 4
})
# Save the mapped columns back into df
df['Study_Habit']      = study_habit_num
df['Study_Group_Freq'] = study_group_num
df['Commute_Time']     = commute_hrs
df['Extracurricular']  = extracurricular_num

print(df.dtypes)
print(df.head())
```

```python
df['Performance_Tier'].unique()
```

```python
performance_tier_map = {
    'At-Risk': 0,
    'Fair / Average': 1,
    'Very Good / Good' : 1,
    'Outstanding / Excellent': 2
}
df['Performance_Tier'] = df['Performance_Tier'].map(performance_tier_map)
```

```python
df.drop(columns=[
    'Self_Study_Hrs',    # → Total_Productive_Hrs, Study_Quality, Study_Distraction_Ratio
    'Tutoring_Hrs_Wk',  # → Total_Productive_Hrs
    'Social_Media_Hrs', # → Distraction_Hrs, Available_Study_Time
    'Gaming_Hrs',       # → Distraction_Hrs, Available_Study_Time
    'Sleep_Hrs',        # → Wellbeing_Score, Available_Study_Time
    'Exercise_Days_Wk', # → Wellbeing_Score
    'Attendance_Rate',  # → Active_Engagement_Score
    'Study_Habit',      # → Study_Quality
    'Study_Group_Freq', # → Active_Engagement_Score
    'Commute_Time',     # → Available_Study_Time
    'Extracurricular',  # → Active_Engagement_Score
], inplace=True)
df
```

```python
from sklearn.feature_selection import SelectKBest, chi2, f_classif 
X = df.drop(['Performance_Tier','GPA'], axis=1)
Y = df['Performance_Tier']
selector = SelectKBest(score_func=f_classif, k=12)
selector.fit(X, Y)
selected_features = selector.get_feature_names_out(input_features=X.columns)
print("Selected Features:", selected_features)
```
