Feature engineer

```python
commute_map = {
    'Under 15 minutes':  0.125,
    '15–30 minutes':     0.375,
    '31–60 minutes':     0.75,
    'More than 1 hour':  1.25   # ← was 'Over 1 hour' before
}
study_group_map = {
    'Never':                               0,
    'Rarely (Before exams only)':          1,
    'Sometimes (Weekly)':                  2,
    'Frequently (Multiple times a week)': 3
}
study_habit_map = {
    'Cramming':        1,
    'Irregular study': 2,
    'Regular study':   3
}
ai_usage_map = { 
    'Never': 0,
    'Rarely': 1,
    'Sometimes': 2,
    'Often': 3,
    'Always': 4
}

# 2. Map directly back into the DataFrame (or keep as standalone Series)
commute_hrs = df['Commute_Time'].map(commute_map)
study_group_num = df['Study_Group_Freq'].map(study_group_map)
study_habit_num = df['Study_Habit'].map(study_habit_map)
extracurricular_num = df['Extracurricular'].map({'Yes': 1, 'No': 0})
ai_tool_usage_num = df['AI_Tool_Usage'].map(ai_usage_map)

# --- Feature 1: Total Productive Hours ---
df['Total_Productive_Hrs'] = df['Self_Study_Hrs'] + df['Tutoring_Hrs_Wk']

# --- Feature 2: Distraction Hours ---
df['Distraction_Hrs'] = df['Social_Media_Hrs'] + df['Gaming_Hrs']

# --- Feature 3: Study-to-Distraction Ratio ---
df['Study_Distraction_Ratio'] = df['Self_Study_Hrs'] / (df['Distraction_Hrs'] + 0.1)

# --- Feature 4: Wellbeing Score ---
df['Wellbeing_Score'] = df['Sleep_Hrs'] + df['Exercise_Days_Wk'] - df['Stress_Level']

# --- Feature 5: Available Study Time ---
df['Available_Study_Time'] = 24 - df['Sleep_Hrs'] - df['Social_Media_Hrs'] - df['Gaming_Hrs'] - commute_hrs

# --- Feature 6: Active Engagement Score ---
df['Active_Engagement_Score'] = df['Attendance_Rate'] / 10 + study_group_num + extracurricular_num

# --- Feature 7: Study Quality Flag ---
df['Study_Quality'] = study_habit_num * df['Self_Study_Hrs']

# --- FIXED Feature 8: AI Enhanced Study ---
# Use the mapped numeric Series (ai_tool_usage_num) instead of the raw text column
df['AI_Enhanced_Study'] = df['Self_Study_Hrs'] * ai_tool_usage_num

# Print results
print(df[['Total_Productive_Hrs', 'Distraction_Hrs', 'Study_Distraction_Ratio',
          'Wellbeing_Score', 'Available_Study_Time', 'Active_Engagement_Score',
          'Study_Quality', 'AI_Enhanced_Study']].head())
```

```python
# Drop high cardinality & leakage
df.drop(columns=['Age', 'University', 'Hometown'], inplace=True)

# Binary
df['Gender']     = df['Gender'].map({'Male': 1, 'Female': 0})
df['STEM_Major'] = df['STEM_Major'].map({'Yes': 1, 'No': 0})

# Ordinal
df['Employment'] = df['Employment'].map({
    'Unemployed / Full-time student': 0,
    'Working part-time':              1,
    'Working full-time':              2
})
df['Diet_Quality'] = df['Diet_Quality'].str.strip().str.lower().map({
    'poor': 1, 'average': 2, 'good': 3, 'excellent': 4
})

df['Internet_Quality'] = df['Internet_Quality'].str.strip().str.lower().map({
    'poor': 1, 'fair': 2, 'good': 3, 'excellent': 4
})
df['AI_Tool_Usage'] = df['AI_Tool_Usage'].str.strip().str.lower().map({
    'never': 0, 'rarely': 1, 'sometimes': 2, 'often': 3, 'always': 4
})
df['Year'] = df['Year'].map({
    'Year 1': 1, 'Year 2': 2, 'Year 3': 3, 'Year 4': 4
})
# Save the mapped columns back into df
df['Study_Habit']      = study_habit_num
df['Study_Group_Freq'] = study_group_num
df['Commute_Time']     = commute_hrs
df['Extracurricular']  = extracurricular_num


print(df.dtypes)
print(df.head())
```

```python
# ============================================================
# ADDITIONAL ENGINEERED FEATURES (No GPA — no leakage)
# ============================================================
# Correlation with Performance_Tier (no GPA):
#   Resource_Quality     +0.603  ★ strongest new feature
#   Lifestyle_Index      +0.587
#   AI_Engagement        +0.533
#   Time_Efficiency      +0.407
#   Digital_Balance      +0.384
#   Academic_Momentum    +0.354
#   Stress_Load          -0.314
#   Study_Consistency    +0.270
#   Stress_Wellbeing     -0.202
# ============================================================

diet_num = df['Diet_Quality']      # already encoded 1-4 from Cell 10
inet_num = df['Internet_Quality']  # already encoded 1-4 from Cell 10

# --- Feature 9: Resource Quality ---
# Diet x Internet synergy — both must be good for effective study
# Strongest new feature: corr +0.603 with Performance_Tier
df['Resource_Quality'] = diet_num * inet_num

# --- Feature 10: Focus Ratio ---
# Total productive hours vs total distraction — measures time discipline
df['Focus_Ratio'] = df['Total_Productive_Hrs'] / (df['Distraction_Hrs'] + 0.1)

# --- Feature 11: Stress-Wellbeing Pressure ---
# High stress on low wellbeing baseline = high academic risk
df['Stress_Wellbeing'] = df['Stress_Level'] / (df['Wellbeing_Score'] + 6)

# --- Feature 12: AI-Powered Engagement ---
# AI tool use x active engagement — tech-enabled learners compound benefits
df['AI_Engagement'] = ai_tool_usage_num * df['Active_Engagement_Score']

# --- Feature 13: Lifestyle Index ---
# Composite of diet, internet, wellbeing — normalized overall life quality
wellbeing_max = df['Wellbeing_Score'].clip(lower=0).max()
df['Lifestyle_Index'] = (
    diet_num / 4 +
    inet_num / 4 +
    df['Wellbeing_Score'].clip(lower=0) / (wellbeing_max + 1e-9)
) / 3

# --- Feature 14: Study Consistency ---
# Study habit strength x group study frequency
# Regular studier who also studies in groups = strongest academic signal
df['Study_Consistency'] = study_habit_num * study_group_num

# --- Feature 15: Time Efficiency ---
# Quality study output relative to stress — high quality under pressure = strong performer
df['Time_Efficiency'] = df['Study_Quality'] / (df['Stress_Level'] + 1)

# --- Feature 16: Digital Balance ---
# AI tool use vs distraction hours — productive vs wasteful screen time ratio
df['Digital_Balance'] = ai_tool_usage_num / (df['Distraction_Hrs'] + 0.1)

# --- Feature 17: Academic Momentum ---
# Active engagement amplified by study habit — showing up + studying consistently
df['Academic_Momentum'] = df['Active_Engagement_Score'] * study_habit_num

# --- Feature 18: Stress Load ---
# Stress x total time burden (distraction + commute) — compound at-risk signal
df['Stress_Load'] = df['Stress_Level'] * (df['Distraction_Hrs'] + commute_hrs)

# ── Verify ───────────────────────────────────────────────────────
new_features = [
    'Resource_Quality', 'Focus_Ratio', 'Stress_Wellbeing',
    'AI_Engagement', 'Lifestyle_Index', 'Study_Consistency',
    'Time_Efficiency', 'Digital_Balance', 'Academic_Momentum', 'Stress_Load'
]
print(f'Added {len(new_features)} new features — no GPA used, no leakage')
print(df[new_features].describe().round(2))
```
